```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
```

```python
#load the data 
data = pd.read_csv("creditcard.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
#check for missing values 
res = data.isnull().sum()
```

```python
#distribution of legit and fraudtrnsactions 
print(data['Class'].value_counts())
```

```python
#seperating the data for analysis 
#0 -  normal transaction 
#1 - fraud transaction 

legit = data[data.Class==0]
fraud = data[data.Class==1] 

print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the both data 

legit_amt = legit.Amount.describe()
print(legit_amt)

fraud_amt = fraud.Amount.describe()
print(fraud_amt)
```

```python
#compare the values for both transaction 

mean = data.groupby("Class").mean()
print(mean)
```

```python
#under samplling 
#sample dataset containing 

legit_sample = legit.sample(n=492)
```

```python
#concatinating two data frames

new_dataset = pd.concat([legit_sample,fraud],axis= 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

```python
#split the data into feature and target 

X = new_dataset.drop(columns="Class",axis=1)
Y = new_dataset['Class']

print(X)
```

```python
print(Y)
```

```python
#split the data into Training  and testing data 

X_train,X_test,Y_train,Y_test = train_test_split(X, Y,test_size=0.2,stratify = Y,random_state=2)



```

```python
#Model
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Accuracy on Trainig data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data',training_data_accuracy)
```

```python
#Accuracy on test data

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Testing data',testing_data_accuracy)
```

```python
#model creation using pickel

f = None
try:
	f = open("creditcard.model","wb")
	pickle.dump(model,f)
except Exception as e:
	print("Issue",str(e))
finally :
	if f is not None:
		f.close()
```
